```python
for index, row in file.iterrows():
    if row["area"] > 3000 and row["bedrooms"] > 4:
        file.drop(index, inplace=True)
```

```python
x=file.iloc[:,1:]
y=file.iloc[:,0:1]
x.head()
```

```python
x["furnishingstatus"]=x["furnishingstatus"].replace("furnished","1")
x["furnishingstatus"]=x["furnishingstatus"].replace("semi-furnished","2")
x["furnishingstatus"]=x["furnishingstatus"].replace("unfurnished","3")
x.head()
```

```python
x["prefarea"]=x["prefarea"].replace("yes","1")
x["prefarea"]=x["prefarea"].replace("no","0")
x.head()
```

```python
x["airconditioning"]=x["airconditioning"].replace("no","0")
x["airconditioning"]=x["airconditioning"].replace("yes","1")
```

```python
x["hotwaterheating"]=x["hotwaterheating"].replace("no","0")
x["hotwaterheating"]=x["hotwaterheating"].replace("yes","1")
x.head()
```

```python
x["basement"]=x["basement"].replace("yes","1")
x["basement"]=x["basement"].replace("no","0")
```

```python
x["guestroom"]=x["guestroom"].replace("yes","1")
x["guestroom"]=x["guestroom"].replace("no","0")
```

```python
x["mainroad"]=x["mainroad"].replace("yes","1")
x["mainroad"]=x["mainroad"].replace("no","0")
x.head()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
lr.score(x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
dr=DecisionTreeRegressor()
dr.fit(x,y)
dr.score(x,y)
```

```python
combined=pd.concat((x,y),axis=1)
```

```python
combined.head()
```

```python
combined.reset_index(drop=True)
combined.to_csv("newHousing.csv")
```

```python
import pickle
pickle.dump(dr,open('model.pkl','wb'))
model = pickle.load(open('model.pkl','rb'))
model.predict([[3000,3,2,2,1,0,0,1,1,1,1,1]])
```
